```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("/world-data-2023.csv")


df.columns = df.columns.str.strip()


print("Available columns:", df.columns.tolist())

birth_df = df[['Country', 'Birth Rate']].dropna()
birth_df = birth_df.sort_values(by='Birth Rate', ascending=False).head(20)

sns.set(style="whitegrid")


colors = sns.color_palette("hsv", len(birth_df))


plt.figure(figsize=(14, 8))
bars = plt.bar(birth_df['Country'], birth_df['Birth Rate'], color=colors)


for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'{height:.2f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')


plt.title("Top 20 Countries by Birth Rate", fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Country", fontsize=14)
plt.ylabel("Birth Rate", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/world-data-2023.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Choose the variable to plot (e.g., Birth Rate)
variable = "Birth Rate"

# Drop missing values
data = df[variable].dropna()

# Set style
sns.set(style="darkgrid")

# Create histogram
plt.figure(figsize=(10,6))
sns.histplot(data, bins=20, kde=True, color="skyblue")

# Customize chart
plt.title(f"Histogram of {variable}", fontsize=16, fontweight="bold", color="darkred")
plt.xlabel(variable, fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.tight_layout()

# Show plot
plt.show()
```
